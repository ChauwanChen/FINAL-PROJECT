# Verificando o número de avaliações e a classificação média para cada livro

```python
query = """
SELECT
    b.book_id,
    b.title,
    COUNT(r.rating_id) AS total_rating,
    ROUND(AVG(r.rating), 2) AS avg_rating
FROM books b
LEFT JOIN ratings r ON b.book_id = r.book_id
GROUP BY b.book_id, b.title
ORDER BY total_rating DESC;
"""
books_avg_rating = pd.read_sql(query, con=engine)
books_avg_rating
```

```python
# Podemos notar o dobro de avaliações para o livro Crepúsculo, que se manteve com uma boa avaliação média
```

# Editoria que lançou o maior número de livros com mais de 50 páginas

```python
query ="""
SELECT
    p.publisher,
    COUNT(b.book_id) AS total_books
FROM books b
JOIN publishers p ON b.publisher_id = p.publisher_id
WHERE b.num_pages > 50
GROUP BY p.publisher
ORDER BY total_books DESC;
"""
filtered_publisher = pd.read_sql(query, con=engine)
filtered_publisher
```

```python
# Penguin books, lançaram 42 livros e foi a editora com maior número de lançamentos.
```

# Autor com a média mais alta de classificações com pelo menos 50 avaliações

```python
query = """
WITH filtered_publisher AS (
    SELECT
        b.book_id,
        b.author_id,
        AVG(r.rating) AS avg_rating,
        COUNT(r.rating_id) AS total_ratings
    FROM books b
    JOIN ratings r ON b.book_id = r.book_id
    GROUP BY b.book_id, b.author_id
    HAVING COUNT(r.rating_id) >= 50
)
SELECT
    a.author,
    ROUND(AVG(b50.avg_rating), 2) AS author_avg_rating
FROM filtered_publisher b50
JOIN authors a ON b50.author_id = a.author_id
GROUP BY a.author
ORDER BY author_avg_rating DESC
LIMIT 1;
"""
df_top_author = pd.read_sql(query, con=engine)
df_top_author
```

```python
# J.K Rowling, tem a melhor avaliação entre os autores 4.28 de 5.0, mais 10 pontos para Grifinória!
```

# Verificando o número médio de avaliações entre os usuários que classificaram mais do que 50 livros

```python
query = """
WITH active_users AS (
    SELECT
        username
    FROM ratings
    GROUP BY username
    HAVING COUNT(DISTINCT book_id) > 50
),
user_reviews AS (
    SELECT
        r.username,
        COUNT(rv.review_id) AS total_reviews
    FROM active_users r
    LEFT JOIN reviews rv ON r.username = rv.username
    GROUP BY r.username
)
SELECT
    ROUND(AVG(total_reviews), 2) AS avg_reviews_per_user
FROM user_reviews;
"""

df_avg_user_reviews = pd.read_sql(query, con=engine)
df_avg_user_reviews
```

```python
# Vemos a média de avaliações entre os usuários que classificaram mais de 50 livros é de cerca de 24 livros!
```

Depois, **a partir desses usuários**, precisamos **contar** o número médio de avaliações, usando a tabela `reviews`. Ou seja, devemos refazer essa última query considerando apenas os usuários filtrados na anterior.
